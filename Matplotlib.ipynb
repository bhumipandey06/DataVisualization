```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Matplotlib Basics

```python
x=np.linspace(1,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("Plot-Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

SubPlots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y**2,x)
plt.subplot(2,2,2)
plt.plot(x**2,y)
```

Object Oriented Way

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.2,0.7,.5,0.3])
axis2.plot(x,y)
```

Types of Plots

Saving plots 

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("1stPlot.png")
```
