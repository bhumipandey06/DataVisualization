```python
import seaborn as sns
import matplotlib.pyplot as plt
```

these plots are used for visualizing distribution of quantitative data(like age, salary, etc)
1.distplot/histplot
2.jointplot
3.pairplot
4.rugplot

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins = 20, kde=True) #displot or histplot both are same
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins = 20, kde=True)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')
```

```python
sns.pairplot(df, hue='size', palette='rainbow')
```
